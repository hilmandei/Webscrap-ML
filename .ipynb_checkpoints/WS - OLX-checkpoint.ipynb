```python
from bs4 import BeautifulSoup
import requests
import joblib
import numpy as np
import pandas as pd
```

```python
listads = joblib.load('listads1')
```

```python
len(listads)
```

```python
df = pd.DataFrame()
```

```python
row = 0
for b in range(0, 3000):
    try:
        url1 = listads[b]
        print(b, url1)
        x = requests.get(url1)
        y = BeautifulSoup(x.content, 'html.parser')
        p = y.find_all("div", class_="detail-extra")[0]
        spec = p.find_all('li')

        harga = int(y.find_all("div", class_="pricelabel tcenter")[0].span.text.replace('.', ''))
        brand = spec[0].a['href'].split('/')[5]

        df.loc[row, 'Brand'] = brand.capitalize()
        df.loc[row, 'Harga'] = harga

        # detail specification
        for i in range(0, len(spec)):        
            df.loc[row, p.ul.find_all('li')[i].text.split('\t')[0].replace(':', '').strip()] = \
            p.ul.find_all('li')[i].text.split('\t')[15].strip()

        row += 1
    except:
        pass
```

```python
df
```
